### Regression tasks

```python
names = [
    'Caco2_Wang',
 'Lipophilicity_AstraZeneca',
 'HydrationFreeEnergy_FreeSolv',
 'Solubility_AqSolDB',
 'LD50_Zhu']
reg_list = []
for name in names:
    all_df = pd.DataFrame()

    trn, val, tst = collect_data(name)
    all_df = trn.merge(val, how='outer')
    all_df = all_df.merge(tst, how='outer')
    # print(name, ':', trn.shape, val.shape, tst.shape, all_df.shape)
    temp_list = all_df[name].tolist()
    new_list = []
    for i in temp_list:
        if i < 200: new_list.append(i)
    reg_list.append(np.array(new_list))
```

```python
group_labels = [name.split('_')[0] for name in names]
fig = ff.create_distplot(reg_list, group_labels, show_hist=False)

fig.update_layout(title_text='Regression task Dataset Distribution')
# fig.update_layout(
#     title_text='Regression task Dataset Distribution',
#     xaxis=dict(type='log', title='X-axis Label'),  # Set type to 'log'
#     yaxis=dict(title='Y-axis Label')  # Add appropriate label for y-axis
# )
fig.show()
```

```python
import plotly.graph_objects as go

def count(df, name):
    zero = df[name].value_counts()[0]
    ones = df[name].value_counts()[1]
    assert len(df) == zero + ones
    print(name, f'\t 0: {zero} | 1: {ones} | total: {zero+ones}')
    return zero, ones
names_cls
```

```python
names = ['CYP2C19_Veith',
 'CYP2D6_Veith',
 'CYP3A4_Veith',
 'CYP1A2_Veith',
 'CYP2C9_Veith',
 'BBB_Martins',
 'Bioavailability_Ma',
 'Pgp_Broccatelli',
 'HIA_Hou',
 'PAMPA_NCATS',
 'hERG_Karim',
 'AMES',

 'DILI',
 'Skin Reaction',
 'Carcinogens_Lagunin',
 'ClinTox']
```

```python
zero_list = []
ones_list = []

name_list = []
type_list = []
count_list = []
for name in names:
    trn, val, tst = collect_data(name)
    all_df = pd.DataFrame()
    all_df = trn.merge(val, how='outer')
    all_df = all_df.merge(tst, how='outer')
    zero, ones = count(all_df, name)
    zero_list.append(zero)
    ones_list.append(ones)

    name = name.split('_')[0]
    name_list.append(name)
    type_list.append('Inactive')
    count_list.append(zero)

    name_list.append(name)
    type_list.append('Active')
    count_list.append(ones)
```

```python
import pandas as pd
dist_df = pd.DataFrame()
dist_df['name'] = pd.DataFrame(name_list)
dist_df['label'] = pd.DataFrame(type_list)
dist_df['count'] = pd.DataFrame(count_list)
```

```python
import plotly.express as px

fig = px.bar(dist_df, x='name', y='count', color='label', text='count')
fig.show()
```
